```python
import pandas as pd
df= pd.read_csv("vins_vinatis_150_pages.csv")
import ast
```

```python
# Ajout du préfixe à la colonne 'image'
df['image_url'] = "https://www.vinatis.com/" + df['image'].astype(str)

# Affichage des 5 premières URLs générées
print(df[['image', 'image_url']].head())

# Optionnel : sauvegarde dans un nouveau CSV
# df.to_csv("vinatis_data_with_image_urls.csv", index=False
```

```python


def safe_literal_eval(val):
    if pd.isna(val):
        return {}
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return {}

def flatten_dict_columns(df, columns):
    for col in columns:
        # Convertir en dict de manière sûre
        df[col] = df[col].apply(safe_literal_eval)
        # Aplatir et préfixer
        flat = pd.json_normalize(df[col]).add_prefix(f'{col}_')
        df = pd.concat([df.drop(columns=[col]), flat], axis=1)
    return df

# Utilisation :
cols_to_flatten = ['features','prices','comment','other_products',]  # Ajoute les colonnes à aplatir
df = flatten_dict_columns(df, cols_to_flatten)
df.head()

```

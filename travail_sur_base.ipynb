```python
import pandas as pd
```

```python
df=pd.read_csv("vins_vinatis_flat_complet.csv")
df.head()
```

```python
colonnes = [
    'id',
    'name',
    'description_short',
    'accroche',
    'image_url',
    'manufacturer_name',
    'contenance',
    'features_abv',
    'features_country',
    'features_region',
    'features_appellation',
    'features_full_grape_variety',
    'features_vintage',
    'features_food_and_wine_matching',
    'features_taste',
    'features_character',
    'features_colour',
    'features_service_temperature',
    'features_type',
    'features_bio',
    'prices_price']
df_select=df[colonnes]
#Colonnes à sélectionner et leurs nouveaux noms
colonnes_fr = {
    'id':'id',
    'name': 'nom',
    'description_short': 'desc',
    'accroche': 'accroche',
    'image_url': 'visuel',
    'manufacturer_name': 'producteur',
    'contenance': 'contenance',
    'features_abv': 'deg_alcool',
    'features_country': 'pays',
    'features_region': 'region',
    'features_appellation': 'appellation',
    'features_full_grape_variety': 'cepages',
    'features_vintage': 'millesime',
    'features_food_and_wine_matching': 'accords',
    'features_taste': 'gout',
    'features_character': 'caractere',
    'features_colour': 'couleur',
    'features_service_temperature': 'temp_serv',
    'features_type': 'type_produit',
    'features_bio': 'bio',
    'prices_price':'prix'}

df_select=df_select.rename(colonnes_fr,axis=1)

df_select.head()
```
